```python
import random
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, n):
        self.weights = np.random.uniform(-1, 1, n)
        self.learning_rate = 0.1

    def feedforward(self, inputs):
        summation = np.dot(inputs, self.weights)
        return self.activate(summation)

    def activate(self, value):
        return 1 if value > 0 else -1

    def train(self, inputs, target):
        guess = self.feedforward(inputs)
        error = target - guess
        self.weights += error * inputs * self.learning_rate

p = Perceptron(2)
training_data = np.array([[0.5, -0.5, 1], [-0.5, 0.5, -1]])

for i in range(len(training_data)):
    inputs = training_data[i][:2]
    target = training_data[i][2]
    p.train(inputs, target)

    prediction = p.feedforward(inputs)
    color = 'g' if prediction == 1 else 'r'
    plt.scatter(inputs[0], inputs[1], c=color)

x1 = -1
y1 = (-p.weights[0] * x1) / p.weights[1]
x2 = 1
y2 = (-p.weights[0] * x2) / p.weights[1]
plt.plot([x1, x2], [y1, y2], c='b')

plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('Perceptron Prediction')
plt.show()
```
